```python
def AB_Test(groupA, groupB, target):
    # Packages 
    from scipy.stats import shapiro # TODO: look up
    import scipy.stats as stats

    # Split A/B
    groupA = groupA[target]
    groupB = groupB[target]

    # Assumption: Normality
    ntA = shapiro(groupA)[1] < 0.05 
    ntB = shapiro(groupB)[1] < 0.05

    # HO: Distribution is Normal! - False
    # H1: Distribution is NOT Normal! - True

    if (ntA == False) & (ntB == False): # H0: Normal Distribution
        # Parametric Test
        # Assumption: Homogeneity of variances
        leveneTest = stats.levene(groupA, groupB)[1] < 0.05
        # H0: Homogeneity: False
        # H1: Heterogeneous: True

        if leveneTest == False:
            # Homogeneity
            ttest = stats.ttest_ind(groupA, groupB, equal_var=True)[1] # standard 2-sample t-test is performed
            # H0: M1 == M2 - False
            # H1: M1 != M2 - True
        else:
            # Heterogeneous
            ttest = stats.ttest_ind(groupA, groupB, equal_var = False)[1] # welch's t-test

            # H0: M1 == M2 - False
            # H1: M1 != M2 - True
    else:
    # Non-Parametric Test
        ttest = stats.mannwhitneyu(groupA, groupB)[1]
        # H0: M1 == M2 - False
        # H1: M1 != M2 - True

    # Result
    temp = pd.DataFrame({
        "AB Hypothesis":[ttest < 0.05],
        "p-value":[ttest]
    })
    temp['Test Type'] = np.where((ntA == False) & (ntB == False), "Parametric", "Non-Parametric")
    temp['AB Hypothesis'] = np.where(temp['AB Hypothesis'] == False, "Fail to Reject H0", "Reject H0")
    temp['Comment'] = np.where(temp['AB Hypothesis'] == "Fail to Reject H0", "A/B groups are similar!", "A/B are not similar!")

    # Columns 
    if (ntA == False) & (ntB == False):
        temp['Homogeneity'] = np.where(leveneTest == False, "Yes", "No")
        print(temp['Homogeneity'])
        temp = temp[["Test Type", "Homogeneity", "AB Hypothesis", "p-value", "Comment"]]
    else:
        temp = temp[["Test Type", "AB Hypothesis", "p-value", "Comment"]]

    # Print Hypothesis
    print('# A/B Testing Hypothesis')
    print('H0: A == B')
    print('H1: A != B', '\n')

    return temp

```

```python
print("People who got 6 and 12 catalogs:")
# Apply A/B Testing
AB_Test(groupA = cats_6, groupB=cats_12, target = 'AmountSpent')

print("People who got 12 and 18 catalogs:")
# Apply A/B Testing
AB_Test(groupA = cats_12, groupB=cats_18, target = 'AmountSpent')

print("People who got 18 and 24 catalogs:")
# Apply A/B Testing
AB_Test(groupA = cats_18, groupB=cats_24, target = 'AmountSpent')
```
